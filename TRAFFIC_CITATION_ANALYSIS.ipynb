```python
#data importing
# path = '/content/drive/MyDrive/Colab/CIS591(Python)/Team_Project/traffic_citations_data.csv'
# df = pd.read_csv(path, low_memory=False)

traffic_df = pd.read_csv('/content/sample_data/traffic_citations_data.csv')
traffic_df.head()
traffic_df.count()
```

```python
# DATA CLEANING
# Changing cleaning data for easy use (Month and Day)

# Creating a mapping for the months
month_mapping = {
    '01-JANUARY': '01', '02-FEBRUARY': '02', '03-MARCH': '03',
    '04-APRIL': '04', '05-MAY': '05', '06-JUNE': '06',
    '07-JULY': '07', '08-AUGUST': '08', '09-SEPTEMBER': '09',
    '10-OCTOBER': '10', '11-NOVEMBER': '11', '12-DECEMBER': '12'
}

# Create a mapping for day of the week
day_mapping = {
    '2-MONDAY': '2', '3-TUESDAY': '3', '4-WEDNESDAY': '4',
    '5-THURSDAY': '5', '6-FRIDAY': '6', '7-SATURDAY': '7', '1-SUNDAY': '1'
}

# Replace the "MONTH" column values using the mapping
traffic_df['MONTH'] = traffic_df['MONTH'].replace(month_mapping)
traffic_df['DAY_OF_WEEK'] = traffic_df['DAY_OF_WEEK'].replace(day_mapping)

# Display the updated DataFrame to verify the changes
traffic_df.tail()
```

## **Question 1**

```python
# Make a bar chart showing the number of tickets for every age

age_df = traffic_without_2024[["SUBJ_AGE", "TICKET"]]


age_ticket = age_df.groupby("SUBJ_AGE")['TICKET'].count().reset_index()

age_fig = px.bar(age_ticket,
              x="SUBJ_AGE",
              y="TICKET",
              title="Total Number of Tickets by Age",
              labels={"SUBJ_AGE": "Age", "TICKET": "Total Ticket Count"},
              width = 800)

age_fig.update_xaxes(tickmode='linear', tick0=15, dtick=5)

age_fig.show()
```

```python
# Make a bar chart showing the number of tickets for Sex ( Male v Female )

sex_df = traffic_without_2024[["SUBJ_SEX", "TICKET"]]


sex_ticket = sex_df.groupby("SUBJ_SEX")['TICKET'].count().reset_index()

sex_ticket_fig = px.bar(sex_ticket,
              x="SUBJ_SEX",
              y="TICKET",
              title="Total Number of Tickets by Sex",
              labels={"SUBJ_SEX": "Sex", "TICKET": "Total Ticket Count"},
              width = 600)

sex_ticket_fig.show()
```

```python
# Make a bar chart showing a breakdown of demographics for civil vs. criminal citation types

def categorize_ticket(row):
    if row["CIVIL_TRAFFIC_VIOLATIONS"] > 0 and row["CRIMINAL_TRAFFIC_VIOLATIONS"] > 0:
        return "Both"
    elif row["CIVIL_TRAFFIC_VIOLATIONS"] > 0:
        return "Civil Only"
    elif row["CRIMINAL_TRAFFIC_VIOLATIONS"] > 0:
        return "Criminal Only"
    else:
        return "None"


traffic_without_2024["TICKET_TYPE"] = traffic_without_2024.apply(categorize_ticket, axis=1)

# Filter relevant columns
demographics_df = traffic_without_2024[["SIMPLE_SUBJ_RE_GRP", "TICKET_TYPE", "TICKET"]]
demographics_df = demographics_df[demographics_df["SIMPLE_SUBJ_RE_GRP"] != "Unknown"]

# Group by race group and ticket type, then count the number of citations
race_ticket_type = demographics_df.groupby(["SIMPLE_SUBJ_RE_GRP", "TICKET_TYPE"])['TICKET'].count().reset_index()

# Create a bar chart
demographics_fig = px.bar(race_ticket_type,
                          x="SIMPLE_SUBJ_RE_GRP",
                          y="TICKET",
                          color="TICKET_TYPE",
                          title="Breakdown of Citation Types by Race Group",
                          barmode="group",
                          labels={"SIMPLE_SUBJ_RE_GRP": "Race Group", "TICKET": "Total Ticket Count", "TICKET_TYPE": "Ticket Type"},
                          width=800)

demographics_fig.show()
```

```python
# Violation Type Breakdown by Race and Sex

# Filter relevant columns
demographics_violations_df = traffic_without_2024[["SIMPLE_SUBJ_RE_GRP", "SUBJ_SEX", "CIVIL_TRAFFIC_VIOLATIONS", "CRIMINAL_TRAFFIC_VIOLATIONS"]]
demographics_violations_df = demographics_violations_df[demographics_violations_df["SIMPLE_SUBJ_RE_GRP"] != "Unknown"]
# Calculate the total count of civil and criminal violations by race and sex
demographics_violations_summary = demographics_violations_df.groupby(["SIMPLE_SUBJ_RE_GRP", "SUBJ_SEX"]).agg(
    civil_count=("CIVIL_TRAFFIC_VIOLATIONS", "sum"),
    criminal_count=("CRIMINAL_TRAFFIC_VIOLATIONS", "sum")
).reset_index()

# Melt the data to make it suitable for stacked bar chart plotting
demographics_violations_melted = demographics_violations_summary.melt(
    id_vars=["SIMPLE_SUBJ_RE_GRP", "SUBJ_SEX"],
    value_vars=["civil_count", "criminal_count"],
    var_name="Violation_Type",
    value_name="Count"
)

# Create a stacked bar chart
demographics_violations_fig = px.bar(demographics_violations_melted,
             x="SIMPLE_SUBJ_RE_GRP",
             y="Count",
             color="Violation_Type",
             facet_col="SUBJ_SEX",
             title="Breakdown of Citation Type by Race Group and Sex",
             labels={"SIMPLE_SUBJ_RE_GRP": "Race Group",
                 "Count": "Total Violations",
                 "Violation_Type": "Violation Type",
                 "SUBJ_SEX": "Sex",
                 },
             barmode="stack",
             width=900)

demographics_violations_fig.show()
```

For further analysis, added Phoenix weather data that includes daily temperatures. Data is form "Visual Crossing Weather "https://www.visualcrossing.com/. Downloaded data from 2018.01 - 2023.12.

```python
import pandas as pd
import plotly.express as px

weather_df = pd.read_csv('/content/sample_data/phoenix_weather.csv')
weather_df.head()

weather_df = weather_df[['datetime', 'temp', 'humidity']].copy()

# Step 1: Convert 'datetime' to datetime format safely (use 'errors="coerce"' to handle invalid entries if needed)
weather_df['datetime'] = pd.to_datetime(weather_df['datetime'], errors='coerce')

# Step 2: Create new columns for 'year', 'month', and 'quarter' using .loc[]
weather_df.loc[:, 'year'] = weather_df['datetime'].dt.year
weather_df.loc[:, 'month'] = weather_df['datetime'].dt.month
weather_df.loc[:, 'quarter'] = weather_df['datetime'].dt.quarter

# Display the updated DataFrame
weather_df = weather_df.drop(columns= ['datetime'])
weather_df.head()

#Change quarter data 1 to Q1, 2 to Q2, 3 to Q3, 4 to Q4

change_quarter = {1: 'Q1', 2: 'Q2', 3: 'Q3', 4: 'Q4'}
weather_df['quarter'] = weather_df['quarter'].replace(change_quarter)

weather_df.head()
```

```python
# Graph (Average tempurate by Year)


year_temp = weather_df.groupby("year")['temp'].mean().reset_index()


fig = px.line(year_temp,
              x="year",
              y="temp",
              title="Average Temperature by Year",
              width = 800)

fig.show()
```

```python
# Graph (Average tempurate by Month)

month_temp = weather_df.groupby("month")['temp'].mean().reset_index()

fig = px.line(month_temp,
              x="month",
              y="temp",
              title="Average Temperature by Month",
              width = 900)

fig.show()
```

```python
# Graph Average temp by Quarter

quarter_temp = weather_df.groupby("quarter")['temp'].mean().reset_index()

fig = px.line(quarter_temp,
              x="quarter",
              y="temp",
              title="Average Temperature by Quarter",
              width = 800)

fig.show()
```

```python
# Trend

monthly_avg_temp = weather_df.groupby(['year', 'month'])['temp'].mean().reset_index()

# Step 4: Create a 'year-month' column for easy plotting
monthly_avg_temp['year_month'] = monthly_avg_temp['year'].astype(str) + '.' + monthly_avg_temp['month'].astype(str).str.zfill(2)

# Step 6: Plot the temperature trend using Plotly Express
fig = px.line(monthly_avg_temp, x='year_month', y='temp',
              title='Monthly Average Temperature from 2018.01 to 2023.12',
              labels={'year_month': 'Year-Month', 'temp': 'Temperature (°C)'},width = 800)

# Customize the layout: Show labels in vertical format and reduce x-ticks for clarity
fig.update_layout(
    xaxis_tickangle=-90,                # Rotate x-axis labels vertically
    xaxis_title='Year-Month',
    yaxis_title='Average Temperature (°C)',
    xaxis=dict(
        tickmode='array',
        tickvals=monthly_avg_temp['year_month'][::2],  # Show every other month to reduce label crowding
        ticktext=monthly_avg_temp['year_month'][::2],  # Use the same values for labeling
    )
)

# Show the plot
fig.show()
```

```python
# check correlatoin between average tempurature and ticket count
# Spearman correlation
# Pearson Correltion is not appropriate for this data, (not showing linear trend)

import pandas as pd
import plotly.express as px

# Assuming 'monthly_avg_temp' has 'month', 'temp' and other data, and 'month_ticket' has 'month' and 'TICKET'

# Step 1: Ensure that the 'month' columns in both DataFrames have the same type
# If not, convert to the same type
monthly_avg_temp['month'] = monthly_avg_temp['month'].astype(int)
month_ticket['MONTH'] = month_ticket['MONTH'].astype(int)

# Step 2: Merge temperature data with ticket data on the 'month' column
merged_df = pd.merge(monthly_avg_temp, month_ticket, left_on='month', right_on='MONTH', how='inner')

# Step 3: Calculate the correlation between temperature and ticket count
correlation = merged_df['temp'].corr(merged_df['TICKET'],method='spearman')
print(f"Spearman correlation between monthly temperature and ticket count: {correlation:.2f}")
print('Pearson Correltion is not appropriate for this data since the data is not in a linear trend')
```
